# **Series**

```python
#1. Series : 시리즈는 인덱스와 값으로 구성된다.

# type 1 : dictionary 자료형과 매우 비슷, zip과 유사
array = pd.Series(['사과','바나나','당근'], index=['a','b','c'])
print(array)
print(array['a'])

# type 2 : dictionary 자료형을 Series로 바꿀 수 있다.
data = {
    'a' : '사과',
    'b' : '바나나',
    'c' : '당근'
}
array = pd.Series(data)
print(array)
print(array['a'])

# type 3: index을 입력하지 않으면 자동으로 된다.
obj = pd.Series([1,2,3,4,5])
print(obj)
```

```python
#2. 값,인덱스,자료형 확인하기
obj = pd.Series([1,2,3,4,5])
print(obj.values)
print(obj.index)
print(obj.dtype)

# 인덱스 변경 가능
obj.index = ['a','b','c','d','e']
print(obj)
```




# **DataFrame**
1.  다수의 시리즈를 모아 처리하기 위한 목적으로 사용
2.  표 형태로 데이터를 손쉽게 출력하고자 할 때 사용

```python
#1. 데이터 프레임 : 데이터 프레임에 들어갈 데이터는 python의 dictionary와 numpy의 array

# type1 : dictionary로 넣기
data = {'fruit': ['apple', 'banana', 'carrot'],
        'price': [100, 200, 300],
        'quantity': [10, 20, 30]}

df = pd.DataFrame(data) # 표 형태로 정리해줘서, 보기 깔끔함
print(df,end='\n\n')

# type2 : numpy로 넣어보기
import numpy as np
array  = np.array([[1,2,3,4,5],[6,8,7,9,10],[7,8,9,10,11]])
df_np = pd.DataFrame(array)
print(df_np)
```

```python
#2. index, columns, values
print(df, end='\n\n')

print(df.index)
print(df.columns)
print(df.values)
```

```python
#3. 각 인덱스에 대해 이름을 짓기가 가능하다.
df.index.name = 'Num'
df.columns.name = 'Info'
print(df)
```

```python
#4. dataframe을 만들면서, colmuns와 index을 설정할 수 있다.
df2 = pd.DataFrame(data, columns = ['fruit','price','quantity'], index = ['one','two','three'])
df2.index.name = 'Num'
df2.columns.name = 'Info'
print(df2)
```

```python
#5. describe : dataframe의 계산 가능한 값들에 대한 다양한 계산 값을 보여준다.
df2.describe()
```

```python
# dataframe - indexing
#6. 열에서 없는 것은 NaN(Not a Number로 출력된다.)
data = {"name" : ["Kilho", "Kilho", "Kilho", "Charles", "Charles"],
        "year" : [2014, 2015, 2016, 2015, 2016],
        "points" : [1.5, 1.7, 3.6, 2.4, 2.9]}

df = pd.DataFrame(data, columns = ["year","name","points","penalty"],
                  index = ["one","two","three","four","five"])
print(df)
```

```python
#7. 열에 접근해서 뽑아내기
print(df['year'], end = '\n\n')
print(df.year, end = '\n\n')

print(df[['year','points']]) # 두 개씩 접근해도 가능하다.
```

```python
#8. NaN에 우리가 원하는 값을 대입할 수 있다.
df['penalty'] = 0.5
print(df,end='\n\n')

df['penalty'] = np.arange(5) # python의 list나 numpy의 array
print(df)
```

```python
#9. 새로운 열을 추가하는 것도 가능하다.
df['zeros'] = np.arange(5)
df
```

```python
#10. 새로운 열에 Series를 추가할 수 도 있다.
# Series로 넣어줄 때 유의해야할 건 index, index를 맞춰서 들어간다.

val = pd.Series([-1.2,-1.5,-1.7], index = ['two', 'four', 'five'])
df['debt'] = val
df
```

```python
#11. numpy의 broadcasting 처럼, 행과 열이 맞다면 쉽게 연산이 가능하다.

df['net_points'] = df['points'] = df['penalty']
df['high_points'] = df['points'] > 2.0
df
```

```python
#12. 열 삭제는 python과 똑같다.
del df['high_points']
del df['net_points']

df
```

```python
#13. dataframe slicing
df[0:3] # 행 기준 0~2까지
```

```python
#14. loc 함수 사용 : 열의 이름과 행의 이름을 써서 접근함
df.loc['two']
```

```python
#15. loc은 다양한 형태로 사용가능하다.
print(df, end = '\n\n')

# df.loc['two':'four', 'points']

# df.loc[:,'year']

# df.loc[:,['year','name']]

# df.loc['three':'five', 'year':'penalty']

df.loc['six',:] = [2013,'Kilho',1.5,0.5,1,2] # 행 추가도 가능하다.

df
```

```python
#16. iloc : index location, 숫자로 location에 접근할 수 있다.
df

# df.iloc[3]
# df.iloc[3:5,0:2]
# df.iloc[[0,1,3],[1,2]]
# df.iloc[:,1:4]
```

```python
# 18. boolean indexing : numpy와 유사하게 boolean indexing도 가능

df
```

```python
#df['year'] > 2014
#df.loc[df['year'] > 2014,:] # loc 에서도 사용가능하다.
#df.loc[df['name']=='Kilho',['name','points']]
#df.loc[(df['points']>2) & (df['points']< 5),:] # 논리연산자도 사용가능하다
df.loc[df['points'] > 3, 'penalty'] = 0 # 새로운 값 대입도 가능하다
df
```

# **Data 수정**

```python
df = pd.DataFrame(np.random.randn(6,4))
df
```

```python
#1. columns와 index 지정하기
df.columns = ['A','B','C','D']

df.index = pd.date_range('20160701',periods=6) # date_range함수는 datetime자료형으로 구성된 날짜 시각등을 알 수 있는 자료형을 만듬

df
```

```python
#2 np.nan은 NaN을 의미한다.
df['F'] = [1.0, np.nan, 3.5, 6.1, np.nan, 7.0]
df
```

```python
#3 NaN없애기 : 행의 값중 하나라도 NaN인 경우 그 행을 없앤다.
df.dropna(how='any')
```

```python
#4 NaN없애기 : 행의 값이 모든 값이 NaN인 경우 그 행을 없앤다
df.dropna(how='all')
```

```python
#5 NaN에 값 넣기
df.fillna(value=0.5)
```

```python
#6. NaN값인지 확인하기
df.isnull()
df.loc[df.isnull()['F'],:] # 다음과 같이 표현할 수도 있다.
```

```python
#7 특정 행 drop하기
df.drop(pd.to_datetime('20160701'))
df.drop([pd.to_datetime('20160701'),pd.to_datetime('20160702')])
```

```python
#8 특정 열 삭제하기
df.drop('F',axis=1)
df.drop(['B','D'],axis=1)
```
